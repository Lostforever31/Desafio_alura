```python
from ipywidgets import IntSlider, Text, Button, Label, HBox, VBox
from IPython.display import display
import google.generativeai as genai
from google.colab import userdata

# Configure o Google Gemini
api_key = userdata.get("SECRET_KEY")
genai.configure(api_key=api_key)

generation_config = {
 "candidate_count": 1,
 "temperature": 0.5,
}

safety_settings = {
 'HATE': 'BLOCK_NONE',
 'HARASSMENT': 'BLOCK_NONE',
 'SEXUAL': 'BLOCK_NONE',
 'DANGEROUS': 'BLOCK_NONE'
}

model = genai.GenerativeModel(
 model_name="gemini-1.0-pro",
 generation_config=generation_config,
 safety_settings=safety_settings
)

chat = model.start_chat(history=[])

# FunÃ§Ã£o para processar os dados
def processar_dados(botao):
    try:
        custo_total = (custo_input.value)
        quanto_gastar = (gastar_input.value)
        tempo_meta = (meta_input.value)
    except ValueError:
        print("Por favor, insira valores numÃ©ricos vÃ¡lidos para os custos e tempo.")
        return

    sonho = sonho_text.value

    if sonho:
        prompt = f"Gostaria de {sonho}, que custa no total {custo_total}, mas consigo gastar apenas {quanto_gastar} e pretendo atingir a meta em {tempo_meta} ."
        response = chat.send_message(prompt)
        print("Resposta:", response.text, "\n")
    else:
        print("Por favor, descreva seu sonho.")

# Widgets
custo_label = Label("Custo total e moeda:")
custo_input = Text(placeholder='Digite o valor')

sonho_label = Label("Qual seu objetivo")
sonho_text = Text(placeholder='Digite sua meta que deseja planejar financeiramente')

# Input para "quanto vocÃª consegue gastar"
gastar_label = Label("Quanto vocÃª consegue gastar em reais ou na moeda desejada:")
gastar_input = Text(placeholder='Digite o valor')

# Input para "quanto tempo pretende atingir a meta"
tempo_label = Label("Em quantos anos/ meses pretende atingir a meta:")
meta_input = Text(placeholder='Digite o tempo ')

# BotÃ£o Ãºnico para enviar
botao_enviar = Button(description='Enviar')
botao_enviar.on_click(processar_dados)

# Organizando a disposiÃ§Ã£o dos elementos com os inputs separados
display(VBox([
  HBox([custo_label, custo_input]),
  HBox([sonho_label, sonho_text]),
  HBox([gastar_label, gastar_input]),
  HBox([tempo_label, meta_input]),
  botao_enviar
]))
```
